## Apply lambda with a function, save answer into a new column
!! Faster than the method below

```python
def multiplyData(A, B):
    answer = A * B
    print(answer)
    return answer
```

```python
df
df['answer'] = df.apply(lambda x: multiplyData(x['A'],x['B']),axis=1); df
```

```python
def multiplyData(A, B):
    answer1 = A * B
    answer2 = A + B
    print(answer1, answer2)
    return [answer1, answer2]
```

```python
df
pd_series = df.apply(lambda x: multiplyData(x['A'],x['B']),axis=1)

pd_series[0]
type(pd_series[0])
```

## DataFrame.apply(user_defined_function)
!! Extremely slow
### applied to `row`, with 1 input argument

```python
def multiplyData(x, y):
    return x * y
```

```python
df
df2 = df.apply(multiplyData, args=[10]); df2
```

### applied to `row`, with 2 input argument

```python
def multiplyData2(x, y, z):
    return x * y / z
```

```python
df
df2 = df.apply(multiplyData2, args=[10, 10]); df2
```

### save results in new columns

```python
def multiplyData3(row, y):
    E = row['A'] * y
    return E
```

```python
df
```

```python
df['E'] = df.apply(multiplyData3 ,axis=1, args=[10]); df
```

```python
# More efficient, since only operating on df['A]

df['E'] = df[['A']].apply(multiplyData3,axis=1, args=[10]); df
```
