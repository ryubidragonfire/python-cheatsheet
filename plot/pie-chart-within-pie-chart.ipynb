## Pie Chart

```python
import matplotlib.pyplot as plt
 
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
def plot_pie_chart(vals, labels):
    import matplotlib.pyplot as plt
    import numpy as np
        
    cmap = plt.get_cmap("tab20c")
    colors = cmap(np.arange(3)*4)
    explode = (0.1, 0, 0)  # explode 1st slice
    
    plt.pie(vals, radius=1, explode=explode, colors=colors, 
               labels=labels, autopct='%1.1f%%')
    
    plt.axis('equal')
    plt.show()
```

```python
vals = np.array([60.,37, 40.])
labels = ['A', 'B', 'C']
plot_pie_chart(vals, labels)
```

## Nested Pie Chart

```python
def plot_nested_pie_chart(vals, size, outer_labels, inner_labels):
    import matplotlib.pyplot as plt
    import numpy as np
    
    fig, ax = plt.subplots()
    
    cmap = plt.get_cmap("tab20c")
    outer_colors = cmap(np.arange(3)*4)
    inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))
    
    _ = ax.pie(vals.sum(axis=1), radius=1,# colors=outer_colors, 
               #labels=outer_labels, #labeldistance=0.90,
               #autopct='%1.1f%%', pctdistance=0.80,
               wedgeprops=dict(width=size, edgecolor='w'))

    _ = ax.pie(vals.flatten(), radius=1-size, colors=inner_colors, 
               labels=inner_labels, #labeldistance=0.9,
               autopct='%1.1f%%', pctdistance=0.7,
               wedgeprops=dict(width=size, edgecolor='w'))

    _ = ax.set(aspect="equal", title='Pie plot with `ax.pie`')
    plt.show()
```

```python
def plot_nested_pie_chart_2(outer_vals, inner_vals, size, outer_labels, inner_labels):
    import matplotlib.pyplot as plt
    import numpy as np
    
    fig, ax = plt.subplots()
    
    cmap = plt.get_cmap("tab20c"); print(cmap)
    outer_colors = cmap(np.arange(3)*4)
    inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))
    
    _ = ax.pie(outer_vals, radius=1, colors=outer_colors, 
               labels=outer_labels, labeldistance=0.90,
               autopct='%1.1f%%', pctdistance=0.80,
               wedgeprops=dict(width=size, edgecolor='w'))

    _ = ax.pie(inner_vals, radius=1-size, colors=inner_colors, 
               labels=inner_labels, labeldistance=0.9,
               autopct='%1.1f%%', pctdistance=0.7,
               wedgeprops=dict(width=size, edgecolor='w'))

    _ = ax.set(aspect="equal", title='Pie plot with `ax.pie`')
    plt.show()
```

```python
size = 0.5
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])
outer_labels = ['blue', 'red', 'green']
inner_labels = ['blue1', 'blue2', 'red1', 'red2', 'green1', 'green2']

plot_nested_pie_chart(vals, size, outer_labels, inner_labels)
```

```python
size = 0.5
outer_vals = np.array([60., 37., 29.])
inner_vals = np.array([10., 50., 10., 27., 10., 10., 9])
outer_labels = ['blue', 'red', 'green']
inner_labels = ['blue1', 'blue2', 'red1', 'red2', 'green1', 'green2', 'green3']

plot_nested_pie_chart_2(outer_vals, inner_vals, size, outer_labels, inner_labels)
```

```python
def plot_nested_pie_chart_3(outer_vals, inner_vals, size, outer_labels, inner_labels):
    import matplotlib.pyplot as plt
    import numpy as np
    
    fig, ax = plt.subplots(figsize=(10,10))
    
    
    cmap = plt.get_cmap("plasma"); 
    outer_colors = cmap(np.arange(len(outer_vals))*round(len(cmap.colors)/len(outer_vals))) #sample from equaly spaced colour map
    inner_colors = cmap(np.array([50, 55, 100, 105, 200, 205, 210]))
    
    _ = ax.pie(outer_vals, radius=1, colors=outer_colors, 
               labels=outer_labels, labeldistance=1.1,
               autopct='%1.1f%%', pctdistance=0.80,
               wedgeprops=dict(width=size, edgecolor='w'))

    print(type(inner_vals), inner_vals)

    _ = ax.pie(inner_vals.flatten(), radius=1-size, colors=inner_colors, 
               labels=inner_labels, labeldistance=0.9,
               autopct='%1.1f%%', pctdistance=0.7,
               wedgeprops=dict(width=size, edgecolor='w'))

    _ = ax.set(aspect="equal", title='Pie plot with `ax.pie`')
    plt.show()
```
