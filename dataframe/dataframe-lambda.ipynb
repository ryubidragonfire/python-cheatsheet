```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(8)
df = pd.DataFrame(np.random.randint(0,10,size=(5, 4)), columns=list('ABCD'))
df
```

## DataFrame.apply(lambda)

```python
df2 = df.apply(lambda x : x + 10); df2
```

```python
df2 = df[['A','C']].apply(lambda x : x + 10); df2
```

## DataFrame.apply(user_defined_function)
### applied to `row`, with 1 input argument

```python
def multiplyData(x, y):
    return x * y
```

```python
df
df2 = df.apply(multiplyData, args=[10]); df2
```

### applied to `row`, with 2 input argument

```python
def multiplyData(x, y, z):
    return x * y / z
```

```python
df
df2 = df.apply(multiplyData, args=[10, 10]); df2
```

### applied to `col`, with 1 input argument

```python
def multiplyData(x, y):
    return x * y
```
